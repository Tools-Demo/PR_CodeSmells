```python
import pandas as pd
import matplotlib.pyplot as plt

def Draw_boxplot():
    df = pd.read_csv('10_fold_results_withHPtuning.csv')
    results = []
    names = ['RandomForest', 'LinearSVC', 'LogisticRegression', 'XGBoost', 'MLP', 'BaggedDT', 'NaiveBayes', 'KNN']
    for name in names:
        model_result = df.loc[df.Model == name]        
        results.append(model_result['Test_Accuracy'].values)

    fig = plt.figure(figsize=(15,8))
#     fig.suptitle('Model Comparison')
    ax = fig.add_subplot(111)
    plt.boxplot(results)
    ax.set_xticklabels(['RF', 'SVM', 'LR', 'XGBoost', 'MLP', 'BaggedDT', 'NB', 'KNN'],fontsize=18)
    plt.yticks(size = 18)
#     ax.set_yticklabels(names,fontsize=15)
#     plt.ylabel('Accuracy', fontsize=20)
    plt.show()

Draw_boxplot()
```
